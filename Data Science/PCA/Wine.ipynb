## Perform Principal component analysis and perform clustering using first 3 principal component scores (both heirarchial and k mean clustering(scree plot or elbow curve) and obtain optimum number of clusters and check whether we have obtained same number of clusters with the original data df (class column we have ignored at the begining who shows it has 3 clusters)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
%matplotlib inline
plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['figure.dpi'] = 200
```

```python
df = pd.read_csv(r"C:\Users\rohit\Downloads\Assignments\data science assignment\PCA\wine.csv")
df.head()
```

```python
df['Type'].value_counts()
```

```python
x = df.iloc[:,1:]
y = df['Type']
```

```python
df.info()
```

## PCA

```python
sc = StandardScaler()

x = sc.fit_transform(x)
x
```

```python
pca = PCA()
pcs = pca.fit_transform(x)
pcs
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=3)
pcs = pca.fit_transform(x)
pcs
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
data = pd.DataFrame(pcs, columns=['PC1', 'PC2', 'PC3'])
data['Type'] = y
data
```

## Hierarchical Clustering using PCA data

```python
data1 = data.iloc[:,:-1]
data1
```

```python
from scipy.cluster import hierarchy
from sklearn.cluster import AgglomerativeClustering
```

```python
lk = hierarchy.linkage(data1, method='complete')
dendo = hierarchy.dendrogram(lk)
```

```python
hc = AgglomerativeClustering(n_clusters=3)
y1 = hc.fit_predict(data1)
y1
```

```python
data1['Clusters'] = y1
data1
```

```python
data1['Clusters'].value_counts()
```

```python
data1.groupby('Clusters').agg('mean')
```
