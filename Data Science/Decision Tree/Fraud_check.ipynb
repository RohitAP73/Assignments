## Use decision trees to prepare a model on fraud data treating those who have taxable_income <= 30000 as "Risky" and others are "Good"

* Data Description :

    Undergrad : person is under graduated or not

    Marital.Status : marital status of a person

    Taxable.Income : Taxable income is the amount of how much tax an individual owes to the government 

    Work Experience : Work experience of an individual person

    Urban : Whether that person belongs to urban area or not

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.metrics import classification_report

%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\rohit\Downloads\Assignments\data science assignment\10-Decision Tree\Fraud_check.csv")
df.rename(columns={'Marital.Status' : 'Marital_Status', 'Taxable.Income' : 'Income',
                   'City.Population' : 'Population','Work.Experience' : 'Experience'}, inplace=True)
df
```

## EDA

```python
df.info()
```

```python
df['Status'] = np.where((df['Income'] <= 30000), 'Risky', 'Good')
df1 = df.drop(columns='Income')
df1
```

```python
oe = OrdinalEncoder()
le = LabelEncoder()

df1.iloc[:,[0,1,4]] = oe.fit_transform(df1.iloc[:,[0,1,4]])
df1['Status'] = le.fit_transform(df1['Status'])
df1
```

## Splitting the data

```python
x = df1.iloc[:,:4]
y = df1['Status']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=1)
```

## Building Model

```python
dt = DecisionTreeClassifier(criterion='gini')
```

```python
dt.fit(xtrain, ytrain)

ypred = dt.predict(xtest)
```

```python
print(classification_report(ytest, ypred))
```

```python
print('Training accuracy: ', dt.score(xtrain,ytrain))
print('Testing Accuracy: ', dt.score(xtest,ytest))
```

## Creating tree diagram

```python
plt.rcParams['figure.figsize'] = (10,10)
plt.rcParams['figure.dpi'] = 250
```

```python
tree.plot_tree(dt)
plt.show()
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'criterion' : ["gini", "entropy"],
    'max_depth' : [4,5,6,7,8,9]
    }
```

```python
grid = GridSearchCV(DecisionTreeClassifier(), param_grid=params)
```

```python
grid.fit(xtrain, ytrain)
```

```python
grid.best_params_
```

## Building Improved Model

```python
dt = DecisionTreeClassifier(criterion='gini', max_depth=4)
```

```python
dt.fit(xtrain, ytrain)

ypred = dt.predict(xtest)
```

```python
print(classification_report(ytest, ypred))
```

```python
print('Training accuracy: ', dt.score(xtrain,ytrain))
print('Testing Accuracy: ', dt.score(xtest,ytest))
```

## Tree diagram

```python
tree.plot_tree(dt, feature_names=x.columns, class_names=df['Status'].unique(), filled=True)
plt.show()
```
